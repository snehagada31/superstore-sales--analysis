```python
import pandas as pd

df = pd.read_csv("train.csv")
df.head()
```

```python
## 3. Dataset Overview
df.info()
df.describe()
```

```python
## 4. Data Cleaning
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
df['Month'] = df['Order Date'].dt.month
df[['Order Date','Month']].head()
```

```python
## 5. Exploratory Data Analysis
```

```python
df.groupby('Region')['Sales'].sum()
```

```python
df.groupby('Category')['Sales'].sum()
```

```python
df.groupby('Segment')['Sales'].sum()
```

```python
df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
```

```python
df.groupby('State')['Sales'].sum().sort_values(ascending=False).head(10)
```

```python
import matplotlib.pyplot as plt

df.groupby('Region')['Sales'].sum().plot(kind='bar')

plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Sales")

plt.show()
```

```python
df.groupby('Category')['Sales'].sum().plot(kind='bar')

plt.title("Sales by Category")
plt.show()
```

```python
df.groupby('State')['Sales'].sum().sort_values(ascending=False).head(5)
```

```python
df.groupby('City')['Sales'].sum().sort_values(ascending=False).head(5)
```

```python
df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(5)
```

```python
df.groupby('Region')['Sales'].sum().plot(kind='bar')
```

```python
monthly_sales = df.groupby('Month')['Sales'].sum()
monthly_sales
```

```python
import matplotlib.pyplot as plt

monthly_sales.plot(kind='line', marker='o')

plt.title("Monthly Sales Trend")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.show()
```

```python
## 6. Key Insights

# • West region generated the highest sales.

# • Technology category has the highest revenue.

# • Canon imageCLASS copier is the top selling product.

# • Sales peak during November while February has the lowest sales.
```

```python
## 7. Correlation Analysis
# This section checks the relationship between month and sales.
```

```python
df[['Sales','Month']].corr()
```

```python
sales_region=df.groupby('Region')['Sales'].sum()
sales_region
```

```python
import matplotlib.pyplot as plt

sales_region.plot(kind='bar')

plt.title("Sales by Region")
plt.xlabel("Region")
plt.ylabel("Total Sales")

plt.show()
```

```python
sales_category = df.groupby('Category')['Sales'].sum()
sales_category
```

```python
import matplotlib.pyplot as plt

sales_category.plot(kind='bar')
plt.title("Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales")
plt.show()
```

```python
top_products = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False).head(10)
top_products
```

```python
top_products.plot(kind='bar')

plt.title("Top 10 Products by Sales")
plt.ylabel("Sales")

plt.show()
```

```python
import seaborn as sns

sns.regplot(x='Month', y='Sales', data=df)

plt.title("Month vs Sales Relationship")
plt.show()
```
